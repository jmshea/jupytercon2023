# JupyterCards Demo

```python
# Import display_flashcards()
from jupytercards import display_flashcards
```

```python
# Make list containing 1 flashcard for JupyterCards:
fc = [
  {"front": "JupyterCards",
   "back": "Interactive library for creating animated flashcards in Jupyter notebooks and Jupyter Book"
  },
]
```

```python
# Display the flashcard
display_flashcards(fc)
```

```python
# Match theme of JupyterCon
display_flashcards(fc, front_colors='jupytercon', back_colors='jupytercon')
```

```python
# Let's add another card

fc2 = [
  {"front": "JupyterCards",
   "back": "Interactive library for creating animated flashcards in Jupyter notebooks and Jupyter book"
  },
  {"front": "JupyterQuiz",
   "back": "Interactive library for creating interactive quizzes in Jupyter notebooks and Jupyter book"
  },
]
```

```python
# and display it
display_flashcards(fc2, front_colors = 'jupytercon', back_colors='jupytercon')
```

## Can use LaTeX for pretty math:

```python
# Add a card with LaTeX
fc3 = [
  {"front": "Fourier Transform",
   "back": "$\int_{-\infty}^{\infty} f(x) e^{-j 2 \pi ft} dt$"
  },
]
```

```python
display_flashcards(fc3, front_colors='jupytercon', back_colors='jupytercon')
```

# Jupyter Quiz Demo

![](figs/mc_schema1.png)

```python
# import display_quiz()
from jupyterquiz import display_quiz 
```

```python
# Add a question with code
q1 = {}
q1['question'] ='"What is the output of the following Python code:'
q1['code'] = 'a="1"\nb="2"\nprint(a+b)'

q1['type'] =  'multiple_choice'
```

```python
# Display it
display_quiz([q1])
```

We need to add some answers:

![](figs/mc_schema2.png)

```python
# Add answers

q1['answers']=[
  {'answer': '3',
   'correct': False,
  },
    {'answer': '12',
   'correct': True,
  }
]
```

```python
# Display it with answers
display_quiz([q1])
```

```python
# Back to RISE for a second
```

```python
# Edit to add feedback:

q1['answers']=[
  {'answer': '3',
   'correct': False,
   'feedback': 'No. Recall that for strings, the + operator performs concatenation!'
  },
    {'answer': '12',
   'correct': True,
  }
]
```

```python
# Display again
display_quiz([q1])
```

```python
# Create a numeric question
nq1 = {
  'question': 'What is the probability of getting exactly 2 heads on 3 flips of a fair coin?',
  'type': 'numeric',
  'answers': 
   [
     {  'type': 'value',
        'value': 0.375,
        'correct': True     
     }
   ]
  }
```

```python
# Display it
display_quiz([nq1])
```

```python
# Copy and add some feedback
numq2 = [
  {'question': 'What is the probability of getting exactly 2 heads on 3 flips of a fair coin?',
   'type': 'numeric',
   'answers': 
   [
     {  'type': 'value',
        'value': 0.375,
        'correct': True     
     },
     
      { 'type': 'range',
        'range': [1.000000000001, 1e12],
        'correct': False,
       'feedback': 'Probabilities cannot be greater than 1'
     }
   ]
  }
]
```

```python
# And display it
display_quiz(numq2)
```
